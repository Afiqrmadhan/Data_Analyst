# PREDIKSI WIN/LOSS DRAFT PICK MOBILE LEGENDS
 

##### Jadi untuk teknisnya, kalau yang win merupakan blue team, berarti value baru di kolom win/loss yaitu : win, kalau yang win berasal dari red team, berarti value baru di kolom win/loss yaitu : loss. Kolom win/loss ini pun dibuat baru untuk mempermudah prediksi

# BERSIHIN DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(color_codes=True)
```

```python
df = pd.read_csv('Mobile Legends Professional League.csv')
```

```python
#Create a new column 'win/lose' based on the conditions
df['win/loss'] = df.apply(lambda row: 'win' if row ['win'] == row['blue'] else 'loss', axis=1)
df.head()
```

### DROP KOLOM-KOLOM YANG TIDAK AKAN DIGUNAKAN SEPERTI HERO YANG DI BAN, WAKTU MATCH, MVP, BLUE AND RED TEAM

```python
df.drop(columns=['date', 'game', 'blue', 'red', 'time', 'win', 'mvp',
                 'ban_blue_1', 'ban_blue_2', 'ban_blue_3', 'ban_blue_4', 'ban_blue_5',
                 'ban_red_1', 'ban_red_2', 'ban_red_3', 'ban_red_4', 'ban_red_5'], inplace=True)
df
```

### TAMPILKAN SEMUA UNIQUE VALUE DARI HASIL DRAFT PICK KEDUA TIM

```python
#Extract unique values from the first to the 10th columns
unique_values_combined = set(df.iloc[:, :10].values.flatten())

#Display the combined unique values
print(list(unique_values_combined))
```

### LAKUKAN LABEL ENCODING UNTUK KE 10 KOLOM DRAFT PRICK HERO

```python
#Manual label encoding mapping
label_encoding_mapping = {value: index + 1 for index, value in enumerate(unique_values_combined)}

#Apply label encoding to all columns
df_encoded = df.apply(lambda col: col.map(label_encoding_mapping) if col.name in df.columns[:10] else col)

#Display the DataFrame with label encoding applied
df_encoded
```

# HASIL PREDIKSI

### HASIL PREDIKSI MENGGUNAKAN DECISION TREE CLASSIFIER

```python
# Import library yang dibutuhkan
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Membagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inisialisasi model Decision Tree
dtree = DecisionTreeClassifier()

# Melatih model menggunakan data latih
dtree.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred = dtree.predict(X_test)

# Menghitung dan menampilkan skor akurasi
print("Accuracy Score :", round(accuracy_score(y_test, y_pred) * 100, 2), "%")
```
